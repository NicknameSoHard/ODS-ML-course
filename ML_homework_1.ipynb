```
import pandas as pd
data = pd.read_csv('drive/Open_course_ML/adult.data.csv')
data.head(10)
```

**1. Сколько мужчин и женщин (признак sex) представлено в этом наборе данных?**

```
data['sex'].value_counts()['Female']
```

**2. Каков средний возраст (признак age) женщин?**

```
data[data['sex'] == 'Female']['age'].mean()
```

**3. Какова доля граждан Германии (признак native-country) в процентах?**

```
(data['native-country'].value_counts()['Germany'] * 100) / data.shape[0]
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак salary) и тех, кто получает менее 50K в год?**

```
print('Для тех, кто получает больше 50К:')
print('Среднее значение возраста: '+ str(data[data['salary'] == '>50K']['age'].mean()) )
print('Среднеквадратическое отклонение (Стандартное отклонение): ' + str(data[data['salary'] == '>50K']['age'].std()) )

print('\nДля тех, кто получает меньше 50К:')
print('Среднее значение возраста: '+ str(data[data['salary'] == '<=50K']['age'].mean()) )
print('Среднеквадратическое отклонение (Стандартное отклонение): ' + str(data[data['salary'] == '<=50K']['age'].std()) )
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters или Doctorate)**

```
# hight_education_list = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
# data[data['salary'] == '>50K' and data['education'] not in hight_education_list].bool()

print('Утверждение ' + str(data[(data['salary'] == '>50K') & ((data['education'] != 'Bachelors') | (data['education'] != 'Prof-school') | (data['education'] != 'Assoc-acdm') | (data['education'] != 'Assoc-voc') | (data['education'] != 'Masters') | (data['education'] != 'Doctorate') )].shape[0] == 0))
```

**7. Выведите статистику возраста для каждой расы (признак race) и каждого пола. Используйте groupby и describe. Найдите таким образом максимальный возраст мужчин расы Amer-Indian-Eskimo.**

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```
count_married = data[(data['salary'] == '>50K') & (data['sex'] != 'Female')  & ((data['marital-status'] == 'Married-civ-spouse') | (data['marital-status'] == 'Married-spouse-absent') | (data['marital-status'] == 'Married-AF-spouse'))].shape[0]
count_not_married = data[(data['salary'] == '>50K') & (data['sex'] != 'Female')  & ((data['marital-status'] != 'Married-civ-spouse') | (data['marital-status'] != 'Married-spouse-absent') | (data['marital-status'] != 'Married-AF-spouse'))].shape[0]
if count_married - count_not_married:
  print('Женатых мужчин больше')
else:
  print('Не женатых мужчин больше')
```

**9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```
print('Максимальная выработка в неделю: ' + str(data['hours-per-week'].max()) )
count_people_that_work_max_hours = data[ data['hours-per-week'] == data['hours-per-week'].max()].shape[0]
print('Людей, которые работают по-максимуму: ' + str(count_people_that_work_max_hours))

print('Людей, которые зарабатывают >50к и работающих максимум времени: ' + str( (data[(data['hours-per-week'] == data['hours-per-week'].max()) & (data['salary'] == '>50K')].shape[0]) / data[ data['hours-per-week'] == data['hours-per-week'].max()].shape[0]))
```
